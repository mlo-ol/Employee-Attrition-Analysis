<h1> Data Cleaning

now that we have explored the dataset, we will now decide how we will clean it. Our target variable is 'Attrition', so we will only keep and modify the features relevant to it. 

<h3> Data cleaning for SQL exploration

```python
# drop the duplicates
df = df.drop_duplicates()

# i decided to use years at the company instead
df = df.drop(columns=["company_tenure_(in_months)"])
```

```python
# to fix Bachelorâ€™s Degree and Masterâ€™s Degree
df['education_level'] = df['education_level'].str.replace("â€™", "'", regex=False)
```

```python
# imputing the missing values with the median
df['distance_from_home'] = df['distance_from_home'].fillna(
    df['distance_from_home'].median()
)
```

```python
# convert the target variable
df['attrition_label'] = df['attrition']  # keep original labels
df['attrition_flag'] = df['attrition'].map({'Stayed': 0, 'Left': 1})  # numeric for calculations
```

```python
# Convert Ordinal Columns Accordingly

# Work Life Balance
df['work_life_balance_label'] = df['work_life_balance']
df['work_life_balance_num'] = df['work_life_balance'].map({
    'Poor': 1, 'Fair': 2, 'Good': 3, 'Excellent': 4
})

# Job Satisfaction
df['job_satisfaction_label'] = df['job_satisfaction']
df['job_satisfaction_num'] = df['job_satisfaction'].map({
    'Low': 1, 'Medium': 2, 'High': 3, 'Very High': 4
})

# Performance Rating
df['performance_rating_label'] = df['performance_rating']
df['performance_rating_num'] = df['performance_rating'].map({
    'Low': 1, 'Below Average': 2, 'Average': 3, 'High': 4
})

# Employee Recognition
df['employee_recognition_label'] = df['employee_recognition']
df['employee_recognition_num'] = df['employee_recognition'].map({
    'Low': 1, 'Medium': 2, 'High': 3, 'Very High': 4
})

# Company Reputation
df['company_reputation_label'] = df['company_reputation']
df['company_reputation_num'] = df['company_reputation'].map({
    'Poor': 1, 'Fair': 2, 'Good': 3, 'Excellent': 4
})

# Job Level
df['job_level_label'] = df['job_level']
df['job_level_num'] = df['job_level'].map({
    'Entry': 1, 'Mid': 2, 'Senior': 3
})
```

```python
# Convert Binary Columns
binary_cols = ['overtime', 'remote_work',
               'leadership_opportunities',
               'innovation_opportunities']

for col in binary_cols:
    df[f'{col}_label'] = df[col]  # keep original labels
    df[f'{col}_flag'] = df[col].map({'No': 0, 'Yes': 1})  # numeric for calculations
```
